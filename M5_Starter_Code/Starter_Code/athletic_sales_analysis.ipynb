```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df_sales_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
df_sales_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
df_sales_2020.head(20)
```

```python
# Display the 2021 sales DataFrame
df_sales_2021.head(20)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.

print(df_sales_2020.dtypes)
```

```python
# Check the 2021 sales data types.
print(df_sales_2021.dtypes)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combine_df = pd.concat([df_sales_2020, df_sales_2021], ignore_index=True)
```

```python
# Check if any values are null.
null_values = combine_df.isnull().sum()
```

```python
# Check the data type of each column
data_types = combine_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combine_df['invoice_date'] = pd.to_datetime(combine_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combine_df
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

rsc_sold = combine_df.groupby(['region','state','city']).agg({'units_sold': 'sum'}).reset_index()
rsc_sold = rsc_sold.rename(columns={'units_sold': 'total_units_sold'})
Region_sales = rsc_sold.sort_values(by='total_units_sold', ascending=False).head(5)
# Show the top 5 results.
print(Region_sales)
```

#### Using `pivot_table`

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
region_sales_total = combine_df.groupby(['region','state','city']).agg({'total_sales': 'sum'}).reset_index()
region_sales_total = region_sales_total.rename(columns={'total_sales': 'total_sales_value'})
top_regions_sales= region_sales_total.sort_values(by='total_sales_value', ascending=False).head(5)


# Show the top 5 results.
print(top_regions_sales)
```

#### Using `pivot_table`

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
topsales_df = combine_df.groupby(['retailer','region','state','city']).agg({'total_sales': 'sum'}).reset_index()
topsales_df = topsales_df.rename(columns={'total_sales': 'Total Sales'})
top_retailers_sales = topsales_df.sort_values(by='Total Sales', ascending=False).head(5)
print(top_retailers_sales)
```

#### Using `pivot_table`

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womenfootwear_df = combine_df[combine_df['product'].str.contains("Women's Athletic Footwear")]
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_retailer_sales = womenfootwear_df.groupby(['retailer','region','state','city']).agg({'units_sold': 'sum'}).reset_index()
women_retailer_sales= women_retailer_sales.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
top_retailers_womens_footwear = women_retailer_sales.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)


# Show the top 5 results.
print( top_retailers_womens_footwear)
```

#### Using `pivot_table`

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
Pivot_sales = pd.pivot_table( combine_df,values='total_sales',index='invoice_date', aggfunc='sum')
decending_value= Pivot_sales.sort_values(by='total_sales', ascending=False).head(10)


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_New = Pivot_sales.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
top_10_daily = daily_New.sort_values(ascending=False).head(10)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
Weekly_New = Pivot_sales.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
top_10_weeks = Weekly_New.head(10)
print(top_10_weeks)
```
